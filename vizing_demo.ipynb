```python
from vizing import *
```

### Example 1
Make a complete graph and then add a new vertex and edge

```python
# make a complete graph of order 4
g = colored_graph(complete = 4)
# add vertex (4)
g.add_vertex()
# connect our new vertex to another vertex
g.connect(4, 0)
# run our edge coloring algorithm
g.color_edges()
# draw our graph
g.draw_graph()

if g.check_correct():
    print("Edge coloring correct!")
else:
    print("Edge coloring incorrect!")
```

### Example 2
Make a complete bigger graph.

```python
g_big = colored_graph(complete = 9)
g_big.color_edges()
g_big.draw_graph()

if g_big.check_correct():
    print("Edge coloring correct!")
else:
    print("Edge coloring incorrect!")
```

```python
matrix = [[0, 1, 1, 1], [1, 0, 1, 1], [1, 1, 0, 1], [1, 1, 1, 0]]
g_matrix = colored_graph(matrix = matrix)
g_matrix.color_edges()
g_matrix.draw_graph()

if g_big.check_correct():
    print("Edge coloring correct!")
else:
    print("Edge coloring incorrect!")
```
